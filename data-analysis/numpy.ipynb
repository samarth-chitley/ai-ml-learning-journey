```python
import numpy as np
```

```python
# Create array using numpy
# 1D Array:
arr1 = np.array([1,2,3,4,5])
```

```python
print(arr1, type(arr1))
print(arr1.shape)
```

Create a new 1D array and reshape it to 2D array

```python
arr2 = np.array([1,2,3,4,5])
arr2.reshape(1,5)
```

.reshape: To re-shape ndarray.
1. In above example arr2 is 1D array.
2. Reshaped array is viewd as having 2 open/close brackets - which signifies that it is a 2D array.
3. Reshaped it to (1,5) -> (row,col)

```python
#Creating 2D Array
arr3 = np.array([[1, 2, 3, 4, 5]])
arr3.shape #(1,5) signifies it has 1 row & 5 col
```

```python
arr4 = np.array([[1, 2, 3, 4, 5],[2, 3, 4, 5, 6]])
print(arr4, arr4.shape, type(arr4), sep="\n")
```
